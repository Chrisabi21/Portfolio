```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.util import ngrams
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Data collection (assuming data is already collected and stored in a CSV file)
data = pd.read_csv('social_media_data.csv')

# Data preprocessing
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    # Tokenization
    words = word_tokenize(text.lower())
    
    # Removing stopwords and stemming
    filtered_words = [stemmer.stem(word) for word in words if word not in stop_words]
    
    return ' '.join(filtered_words)

data['processed_text'] = data['text'].apply(preprocess_text)

# Sentiment labeling with VADER
analyzer = SentimentIntensityAnalyzer()

def vader_sentiment_label(text):
    compound_score = analyzer.polarity_scores(text)['compound']
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

data['vader_sentiment'] = data['processed_text'].apply(vader_sentiment_label)

# Feature extraction with Bag-of-Words (BoW)
vectorizer = CountVectorizer(ngram_range=(1, 3))
X_bow = vectorizer.fit_transform(data['processed_text'])
y = data['vader_sentiment']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=42)

# Model development: Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
nb_predictions = nb_classifier.predict(X_test)

# Model development: Logistic Regression
lr_classifier = LogisticRegression(max_iter=1000)
lr_classifier.fit(X_train, y_train)
lr_predictions = lr_classifier.predict(X_test)

# Model evaluation
print("Naive Bayes Classifier:")
print("Accuracy:", accuracy_score(y_test, nb_predictions))
print("Classification Report:\n", classification_report(y_test, nb_predictions))

print("\nLogistic Regression Classifier:")
print("Accuracy:", accuracy_score(y_test, lr_predictions))
print("Classification Report:\n", classification_report(y_test, lr_predictions))

# Insight generation
feature_names = vectorizer.get_feature_names_out()
coefficients = lr_classifier.coef_

top_positive_features = [feature_names[idx] for idx in coefficients.argsort()[0][-10:]]
top_negative_features = [feature_names[idx] for idx in coefficients.argsort()[0][:10]]

print("\nTop 10 Positive Features:")
print(top_positive_features)

print("\nTop 10 Negative Features:")
print(top_negative_features)
```

```python
import matplotlib.pyplot as plt

def level_of_agreement(company):
    ground_truth_labels = company['ground_truth_sentiment']
    predicted_sentiments = company['lexicon_sentiment'] 
    
    agreement_count = sum(1 for true_label, lexicon_label in zip(ground_truth_labels, predicted_sentiments) if true_label == lexicon_label)
    disagreement_count = len(ground_truth_labels) - agreement_count
    
    agreement_counts = {'Agreement': agreement_count, 'Disagreement': disagreement_count}

    return agreement_counts

# Example data for three companies
company1 = {'ground_truth_sentiment': ['positive', 'negative', 'positive', 'neutral', 'negative', 'positive'],
            'lexicon_sentiment': ['positive', 'negative', 'positive', 'neutral', 'positive', 'positive']}
company2 = {'ground_truth_sentiment': ['positive', 'positive', 'neutral', 'neutral', 'negative', 'positive'],
            'lexicon_sentiment': ['positive', 'positive', 'neutral', 'negative', 'positive', 'negative']}
company3 = {'ground_truth_sentiment': ['negative', 'positive', 'neutral', 'neutral', 'negative', 'positive'],
            'lexicon_sentiment': ['negative', 'positive', 'neutral', 'negative', 'negative', 'positive']}

# Create subplots for three companies
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot pie charts for each company
for i, company in enumerate([company1, company2, company3]):
    agreement_counts = level_of_agreement(company)
    axs[i].pie(agreement_counts.values(), labels=agreement_counts.keys(), autopct='%1.1f%%', explode=(0.1, 0), shadow=True, startangle=140, colors=['skyblue', 'lightcoral'])
    axs[i].set_title(f'Company {i+1} Agreement and Disagreement')

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Function to train and evaluate a classifier
def train_and_evaluate_classifier(X_train, X_test, y_train, y_test, vectorizer, classifier):
    # Vectorize the text data
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
    
    # Train the classifier
    classifier.fit(X_train_vec, y_train)
    
    # Predict labels for the test set
    y_pred = classifier.predict(X_test_vec)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    return accuracy

# Example data for one company
company_data = {
    'text': ['text1', 'text2'],
    'sentiment': ['positive', 'negative']
}

# Define classifiers
classifiers = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=1000)
}

# Define vectorizers
vectorizers = {
    'Bag-of-Words': CountVectorizer(),
    'TF-IDF': TfidfVectorizer()
}

# Number of iterations
n_iterations = 10

# Results dictionary to store accuracies
results = {}

# Perform multiple runs
for classifier_name, classifier in classifiers.items():
    for vectorizer_name, vectorizer in vectorizers.items():
        for i in range(n_iterations):
            # Split the data into training and testing sets
            X_train, X_test, y_train, y_test = train_test_split(company_data['text'], company_data['sentiment'], test_size=0.2, random_state=i)
            
            # Train and evaluate the classifier
            accuracy = train_and_evaluate_classifier(X_train, X_test, y_train, y_test, vectorizer, classifier)
            
            # Store the accuracy
            key = (classifier_name, vectorizer_name)
            if key not in results:
                results[key] = []
            results[key].append(accuracy)

# Calculate mean and standard deviation of accuracies
summary = {}
for key, accuracies in results.items():
    mean_accuracy = np.mean(accuracies)
    std_accuracy = np.std(accuracies)
    summary[key] = {'Mean Accuracy': mean_accuracy, 'Std Deviation': std_accuracy}

# Create DataFrame from summary
summary_df = pd.DataFrame.from_dict(summary, orient='index')

# Plot results (optional)
summary_df.plot(kind='bar', y='Mean Accuracy', yerr='Std Deviation', figsize=(10, 6), legend=None, title='Mean Accuracy and Standard Deviation')
plt.xlabel('Classifier - Vectorizer Combination')
plt.ylabel('Mean Accuracy')
plt.xticks(rotation=45)
plt.show()

# Display summary
print(summary_df)
```
