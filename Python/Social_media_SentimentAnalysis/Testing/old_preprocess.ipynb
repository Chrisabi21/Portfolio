```python
import pandas as pd
import re

from langdetect import detect
from tqdm import tqdm

from nltk.tokenize import word_tokenize
from nltk.tokenize import WhitespaceTokenizer
from nltk.tokenize import WordPunctTokenizer
from nltk.tokenize import TreebankWordTokenizer
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import TweetTokenizer
```

```python
projectData = pd.read_csv("02.csv", lineterminator='\n')
projectData.shape
```

```python
def check_language(text):
    # Check if tweet is in English language
    try:
        if detect(text) == 'en':
            return text
    except:
        return None
```

```python
# Function to clean data   
def clean_tweet(tweet):
    # text case normalization
    tweet = tweet.lower()
    
    # Remove special characters, numbers, links,username, hashtags, and punctuation
    # Remove usernames and mentions   
    cleaned_tweet = re.sub(r'@\w+', '', tweet)
    
    # Remove links 
    cleaned_tweet = re.sub(r'http[s]?://\S+', '', tweet)
    
    # Remove hashtags 
    cleaned_tweet = re.sub(r'#\w+', '', cleaned_tweet)
    
    # Remove Characters
    cleaned_tweet = re.sub(r'[^a-zA-Z\s]', '', cleaned_tweet)
    
    # Remove extra spaces
    cleaned_tweet = re.sub(r'\s+', ' ', cleaned_tweet).strip()
    
    return cleaned_tweet
```

```python
# Clean the 'tweet' column using the clean_tweet function
cleaned_dfs = []
for i in tqdm(range(0, len(projectData), 500000)):
    batch_df = projectData.loc[i:i+500000].copy()
    batch_df['cleaned_tweet'] = batch_df['text'].apply(check_language)
    batch_df.dropna(subset=['cleaned_tweet'], inplace=True)
    cleaned_dfs.append(batch_df)

# Combine cleaned DataFrames
cleaned_df = pd.concat(cleaned_dfs)

# Display the cleaned DataFrame
print(cleaned_df.shape)
cleaned_df.head()
```

```python
cleaned_df.to_csv("01.csv")
```
