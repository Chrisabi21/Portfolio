PCA FEATURE EXTRACTION

```python
!pip install opencv-python
!pip install plotly
import os
import numpy as np
# load and show an image with Pillow
from PIL import Image
import cv2
from matplotlib.pyplot import imshow
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import plotly.express as px
import plotly.io as pio
import random
pio.renderers.default = "svg"
```

```python
N_COMPONENTS = 3
NUM_IMAGES = 500 # Random number of images used for clustering
NUMBER_OF_CLUSTERS = 8

# The Path to the image dataset
image_dataset_path = "./ocular-disease-recognition-odir5k/preprocessed_images/"
```

```python
#function that returns numpy array of the images in the folder
def read_images(image_folder_path):
    image_files = [os.path.join(image_folder_path, fname) for fname in os.listdir(image_folder_path) if fname.endswith('.jpg')]
    random.shuffle(image_files)

    for i in range(0, NUM_IMAGES):
        yield cv2.imread(image_files[i], cv2.IMREAD_UNCHANGED)
```

```python
data = None
for image_batch in read_images(image_dataset_path):
    flatten = image_batch.reshape(-1)
    if data is None:
        data = flatten
    else:
        data = np.vstack([data, flatten])
```

```python
# View first image
first_image = data[:1][0]
first_image.shape
first_image_array = first_image.reshape(512, 512, 3) # reshape to image from flattened
PIL_image = Image.fromarray(first_image_array).convert('RGB')
imshow(np.asarray(PIL_image))
```

```python
# PCA Dimension reduction
pca = PCA(n_components=N_COMPONENTS)
reduced_data = pca.fit_transform(data)
```

```python
# Variance ration
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)

px.area(
    x=range(1, exp_var_cumul.shape[0] + 1),
    y=exp_var_cumul,
    labels={"x": "# Components", "y": "Explained Variance"}
)
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(NUMBER_OF_CLUSTERS).fit(reduced_data)
```

```python
clustering.labels_
```

```python
# Visualisation of pca charts, with color representing labels
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}
fig = px.scatter_matrix(
    reduced_data,
    labels=labels,
    dimensions=range(N_COMPONENTS),
    color=clustering.labels_
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
