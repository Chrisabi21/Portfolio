```python
import os
import numpy as np
import pandas as pd
from skimage.io import imread
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

```

```python
BATCH_SIZE = 10
N_COMPONENTS = 5
CLUSTERS = 5

# Path to your image dataset
 
image_dataset_path = "./ocular-disease-recognition-odir5k/preprocessed_images/"


```

```python
def read_images_in_batches(image_folder_path, batch_size=BATCH_SIZE):
    image_files = [os.path.join(image_folder_path, fname) for fname in os.listdir(image_folder_path) if fname.endswith('.jpg')]
    for i in range(0, len(image_files), batch_size):
        yield [imread(f) for f in image_files[i:i + batch_size]]
```

```python
def preprocess_images(image_batch):
    # Convert images to flattened arrays and stack them
    data = [img.flatten() for img in image_batch]
    return np.vstack(data)

```

```python
# Read, preprocess and apply PCA in batches
all_pca_data = []

for image_batch in read_images_in_batches(image_dataset_path):
    data = preprocess_images(image_batch)
    scaler = StandardScaler()
    data = scaler.fit_transform(data)
    
    pca = PCA(n_components=N_COMPONENTS)
    pca_data = pca.fit_transform(data)
    
    all_pca_data.append(pca_data)

```

```python
all_pca_data = np.vstack(all_pca_data)

# Apply clustering on PCA reduced data
kmeans = KMeans(n_clusters=CLUSTERS)
kmeans.fit(all_pca_data)

# Print cluster assignments (if needed)
print(kmeans.labels_)
```
