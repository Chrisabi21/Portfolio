```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
```

```python

def read_images_in_batch(image_folder, start, end):
    image_files = sorted(os.listdir(image_folder))[start:end]
    batch_images = [cv2.imread(os.path.join(image_folder, img), cv2.IMREAD_GRAYSCALE) for img in image_files]
    flattened_images = [img.flatten() for img in batch_images]
    return np.vstack(flattened_images)
```

```python

# Folder containing images
image_folder =  "./ocular-disease-recognition-odir5k/preprocessed_images/"

# Check if the folder exists
if not os.path.exists(image_folder):
    print("The folder does not exist.")
    exit()

# Read image filenames and set batch size
all_image_files = sorted(os.listdir(image_folder))
batch_size = 2
n_batches = len(all_image_files) // batch_size



```

```python
# Initialize X with shape (0, n_features)
sample_img = cv2.imread(os.path.join(image_folder, all_image_files[0]), cv2.IMREAD_GRAYSCALE)
n_features = sample_img.size
X = np.empty((0, n_features))

# Read images and update data matrix X batch-wise
for i in range(n_batches):
    start = i * batch_size
    end = (i + 1) * batch_size
    X_batch = read_images_in_batch(image_folder, start, end)
    X = np.vstack((X, X_batch))
```

```python
# Perform FastICA
n_components = 4
ica = FastICA(n_components=n_components, max_iter=500, tol=1e-4)
X_ica = ica.fit_transform(X)
A_ica = ica.mixing_

# Reshape to original image shape
image_shape = sample_img.shape
```

```python

# Display images
plt.figure(figsize=(12, 6))

# Show original images (first batch for demonstration)
for i in range(batch_size):
    img = cv2.imread(os.path.join(image_folder, all_image_files[i]), cv2.IMREAD_GRAYSCALE)
    plt.subplot(2, batch_size, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f'Original {i + 1}')
    plt.axis('off')

```

```python

# Show ICA components
for i in range(n_components):
    plt.subplot(2, batch_size, batch_size + i + 1)
    plt.imshow(A_ica[:, i].reshape(image_shape), cmap='gray')
    plt.title(f'ICA {i + 1}')
    plt.axis('off')

plt.show()
```
