## Age distribution 
  Here below we are creating a grid of histograms one for each label (N,D,G,C,A,H,M,O), to show age distribution. 
  This would help us compare the age profiles across different labels easily.
  we are doing this again because the data has been split based on each eye (right and left eye) and diagnosis of an individual.

```python
# List of columns for to plotting our age distribution
label_columns = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']

# Initialize the Matplotlib figure
plt.figure(figsize=(20, 15))

# Loop through each label column and create a subplot for its histogram
for i, label in enumerate(label_columns):
    plt.subplot(3, 3, i + 1) #grid dimensions
    
    # Filtering rows where the label is set, then plotting the  histogram of ages for those rows
    sns.histplot(original_dataframe[original_dataframe[label] == 1]['Patient Age'], bins=20, kde=False)
    plt.title(f'Age Distribution for {label}')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Analysis : We carried out this visulization to understand the distribution of ages for each label on the data set 
though we had already done this, we are doing it again because at this stage, the data has been processed ( split into right eye and diagnotic and left eye and diagnostic) and we expect a better level of accuracy in the distribution for each label. which is shown by comparing the previous dataset visulaization titled disease and age distribution against this one.

```python
# List of columns for to plotting our age distribution
label_columns = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']

# Initialize the Matplotlib figure
plt.figure(figsize=(20, 15))

# Loop through each label column and create a subplot for its histogram
for i, label in enumerate(label_columns):
    plt.subplot(3, 3, i + 1) #grid dimensions
    
    # Filtering rows where the label is set, then plotting the  histogram of ages for those rows
    sns.histplot(original_dataframe[original_dataframe[label] == 1]['Patient Age'], bins=10, kde=False)
    plt.title(f'Age Distribution for {label}')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
